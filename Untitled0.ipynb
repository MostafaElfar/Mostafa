```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import model_selection
import glob
import tqdm
import random
```

```python
def pair(path):
  files_list= glob.glob(path+"/*.jpg")
  lst_1=[]
  lst_2 = []
  target = []

  for path in tqdm.tqdm(files_list):
    lst_1.append(path)
    lst_2.append(path)
    target.append(1)

    while True:

      i = np.random.randint(0, len(files_list))
      if path !=  files_list[i]:
        lst_1.append(path)
        lst_2.append(files_list[i] )
        target.append(0)
        break

  data_dic={
      "image_1":lst_1,
      "image_2": lst_2,
      "label": target
  }
  df = pd.DataFrame(data_dic)
  return df

```

```python
df = pair("/content/group1/group1")
```

```python
train ,test = model_selection.train_test_split(df , random_state=42,test_size=0.1 , stratify=df.label)
```

```python
def get_tf(x,y):
  x1,x2 = x

  x1 = tf.io.read_file(x1)
  x1 = tf.image.decode_jpeg(x1)
  x1 = tf.image.convert_image_dtype(x1 ,tf.float32)
  x1 = tf.image.resize(x1,(244,244))
  x1 = tf.clip_by_value(x1  , 0.0 ,0.1)

  x2 = tf.io.read_file(x2)
  x2 = tf.image.decode_jpeg(x2)
  x2 = tf.image.convert_image_dtype(x2 ,tf.float32)
  x2 = tf.image.resize(x2,(244,244))
  x2 = tf.clip_by_value(x2  , 0.0 ,0.1)

  return (x1,x2), y
```

```python
def augmant( x , y):
  x1, x2 = x
  x1 = tf.image.random_flip_left_right(x1)
  x1 = tf.image.random_flip_up_down(x1)
  x1 = tf.image.random_brightness(x1 , max_delta=32/255)
  x1 = tf.image.random_saturation(x1 , lower=0.5 , upper= 1.5)

  x2 = tf.image.random_flip_left_right(x2)
  x2 = tf.image.random_flip_up_down(x2)
  x2 = tf.image.random_brightness(x2 , max_delta=32/255)
  x2 = tf.image.random_saturation(x2 , lower=0.5 , upper= 1.5)


  return (x1,x2), y
```

```python
def mds( x1,x2 , y ,training=False ):

    data = tf.data.Dataset.from_tensor_slices(((x1,x2),y))
    data = data.shuffle(2000)

    data = data.map(get_tf,num_parallel_calls=tf.data.experimental.AUTOTUNE)

    if training:
      data = data.map(augmant,num_parallel_calls=tf.data.experimental.AUTOTUNE)
    data = data.batch(batch_size=32)
    data = data.prefetch(tf.data.experimental.AUTOTUNE)

    return data
ds_train = mds(train.image_1, train.image_2, train.label ,training=True)
ds_test= mds(test.image_1, test.image_2, test.label )
```

```python
cnn = tf.keras.applications.DenseNet121(
    include_top=False,
    input_shape=(244,244,3),

)
cnn.trainable = False
cnn.summary()
```

```python
class Rec(tf.keras.Model):
  def __init__(self,cnn):
    super(Rec, self).__init__()

    self.cnn = cnn
    self.d1 = tf.keras.layers.Dense(512, use_bias=False )
    self.bn1 = tf.keras.layers.BatchNormalization()
    self.d2 = tf.keras.layers.Dense(128, use_bias=False , )
    self.bn2 = tf.keras.layers.BatchNormalization()
    self.out = tf.keras.layers.Dense(units=1, activation=tf.keras.activations.sigmoid)

  def call(self, input):
    x1, x2 = input

    x1 = self.cnn(x1)
    x1 = tf.keras.layers.GlobalAveragePooling2D()(x1)
    x2 = self.cnn(x2)
    x2 = tf.keras.layers.GlobalAveragePooling2D()(x2)

    x= tf.keras.layers.concatenate([x1,x2])
    x = self.d1(x)
    x = self.bn1(x)
    x= tf.nn.leaky_relu(x)

    x = self.d2(x)
    x = self.bn2(x)
    x= tf.nn.leaky_relu(x)
    output = self.out(x)

    return output
  def build(self, dim=[244,244,3]):
   x = (tf.keras.layers.Input(dim) , tf.keras.layers.Input(dim))
   return tf.keras.Model(inputs=[x],outputs=[self.call(x)])
model = Rec(cnn)
model = model.build()
model.summary()
```

```python
tf.keras.utils.plot_model(model)
```

```python
model.compile(optimizer=tf.optimizers.Adam(),
              loss=tf.losses.BinaryCrossentropy(),
              metrics=["acc"])
model.fit(ds_train,epochs=8,validation_data=ds_test,batch_size=32 )
```
